```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
q = requests.get('https://en.wikipedia.org/wiki/FTSE_100_Index')
soup1 = BeautifulSoup(q.text,'lxml')
soup1
```

```python
table=soup1.find('table',{'id':'constituents'}).find('tbody')
table
```

```python
table=soup1.find('table',{'id':'constituents'}).find('tbody').findAll('tr')[1:]
table[7]
```

```python
df_scraped = pd.DataFrame()
for row in table:
        title = row.find('a')['title']
        ticker = row.findAll('td')[1].text
        industry = row.findAll('td')[2].text.strip()
        row1 = pd.Series({"Ticker":ticker,"Industry":industry},name=title)
        df_scraped = pd.concat([df_scraped, row1], axis=1)
df_scraped = df_scraped.T
```
